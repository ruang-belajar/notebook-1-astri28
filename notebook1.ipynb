```python

# PERSIAPAN DATA

# download file contoh csv
!wget https://github.com/ruang-belajar/python/blob/master/notebook/data/2013-data-kebakaran.csv

import pandas as pd

# load data ke dataframe (df)
df = pd.read_csv("2013-data-kebakaran.csv")
df.head()
```

Filter data laporan kebakaran yang sumber informasinya dari Masyarakat

```python
filtered_data = df[df['sumber_info'] == 'Masyarakat']
display(filtered_data)
```

 Filter data laporan kebakaran yang sumber informasinya dari Masyarakat di bulan Februari


```python
filtered_data = df.query('sumber_info == "Masyarakat" and bulan == 2')
display(filtered_data)
```

 Filter data laporan kebakaran dengan nilai kerugian lebih dari 100juta




```python
filtered_data = df[df['kerugian'] == 100000000]
display(filtered_data)
```

Filter data laporan kebakaran selama semester 2 juli - desember


```python
filtered_data = df[df['bulan'].between(7, 12)]
display(filtered_data)
```
